6. What are the weaknesses of grid search? Why wouldn't we want to use it?
  - The first weakness of the grid search is the time required to approximate the best set of parameters. Without any heuristic used, every possible combination of parameters must be iterated through and tested. This process is much more time consuming that search methods that use a heuristic in order to only test parameter values that trend towards a lower amount of error. Another weakness of grid search is that it is limited to the resolution of grid points searched. If points used in the grid search are only whole values, and the true optimum set of parameters lie on half values, then that true optimum will not be found, and instead only a relative optimum will be discovered. Lastly, a weakness of grid search is that it limits the range of values tested. If parameter values between 0 and 10 are only tested, but the optimum values actually occur in a range of values greater than 10, grid search does not find this set of points and instead only finds the approximate best set of parameters within the specified range.

## Experiment 1: Coarse Grid Search - Gaussian Distribution 
### Loading gaussdist.csv:

```python
data = np.loadtxt(open("gaussdist.csv", "rb"), delimiter=",")
data[:5]
```

- #### Identifying response variable and feature variable:
  - Based from the Gaussian Distribution data that is loaded in the above cell, it is clear that the second column is not monotonic, which is a characteristic of the output for a Gaussian Distribution. As such, the first column are the feature variable values, and the second column are the response variable values.

### Instantiating GaussianCostFunction using the provided data:

```python
features = data[:, 0]
responses = data[:, 1]

gaussian_func = GaussianCostFunction(features, responses)
```

### Plotting the provided data:

```python
plt.scatter(features, responses, label='feature and true response point')
plt.title("Response Values vs. Feature Values", fontsize=15)
plt.xlabel("Feature Values", fontsize=12)
plt.ylabel("Response Values", fontsize=12)
plt.suptitle('Original Data', fontsize=18)
plt.legend(loc=2)
plt.ylim(-.02, .4)
plt.tight_layout()
```

### Plotting the provided data alongside predicted output:

```python
pred_y = gaussian_func.predict(features, np.array([1, 0.75]))
mse_score = gaussian_func.cost(np.array([1, 0.75]))

plt.scatter(features, responses, label='feature and true response point')
plt.scatter(features, pred_y, label='feature and predicted response point')
plt.title("Response Values vs. Feature Values", fontsize=15)
plt.xlabel("Feature Values", fontsize=12)
plt.ylabel("Response Values", fontsize=12)
plt.suptitle(f'Original Data and Predicted Response (MSE = {round(mse_score, 4)})', fontsize=18)
plt.legend(loc=2)
plt.ylim(-.02, .7)
plt.tight_layout()
```

### Grid Search - Coarse:

```python
mu = np.linspace(0, 10, 100)
sigma = np.linspace(0.5, 2, 50)
mus, sigmas = np.meshgrid(mu, sigma)
params = np.array([mus.flatten(), sigmas.flatten()]).T
mse = np.zeros(params[:, 0].shape)

for i in range(mse.size):
    mse[i] = gaussian_func.cost(params[i])
```

### Heatmap of MSE error for coarse grid search:

```python
df = pd.DataFrame(data=mse.reshape(mus.shape), columns=mu.round(3), index=sigma.round(3))
sns.heatmap(df, cbar_kws={'label':'MSE Error'})
plt.xlabel("Mu Values")
plt.ylabel("Sigma Values")
plt.title("Sigma vs Mu MSE Error")
```

### Visually selecting optimized parameters:

```python
chosen_params = np.array([5, 1.25])
mse_score = gaussian_func.cost(chosen_params)
print(f'Error for mu=5, sigma=1.8: {mse_score}')
```

### Plotting provided data alongside predicted output for selected parameters:

```python
pred_y = gaussian_func.predict(features, chosen_params)

plt.scatter(features, responses, label='feature and true response point')
plt.scatter(features, pred_y, label='feature and predicted response point')
plt.title("Response Values vs. Feature Values", fontsize=15)
plt.xlabel("Feature Values", fontsize=12)
plt.ylabel("Response Values", fontsize=12)
plt.legend()

plt.suptitle(f'Original Data and Predicted Response vars (MSE = {round(mse_score, 4)})', fontsize=18)
plt.legend(loc=2)
plt.ylim(-.02, .45)
plt.tight_layout()
```

## Experiment 2: Refined Grid Search - Gaussian Distribution
### Grid Search - Refinement Pass

```python
mu = np.linspace(5, 6, 100)
sigma = np.linspace(1, 1.75, 50)
mus, sigmas = np.meshgrid(mu, sigma)
params = np.array([mus.flatten(), sigmas.flatten()]).T
mse = np.zeros(params[:, 0].shape)

for i in range(mse.size):
    mse[i] = gaussian_func.cost(params[i])
```

### Heatmap of MSE error for refinement pass grid search:

```python
df = pd.DataFrame(data=mse.reshape(mus.shape), columns=mu.round(3), index=sigma.round(3))
sns.heatmap(df, cbar_kws={'label':'MSE Error'})
plt.xlabel("Mu Values")
plt.ylabel("Sigma Values")
plt.title("Sigma vs Mu MSE Error")
```

### Visually selecting optimized parameters:

```python
chosen_params = np.array([5.485, 1.276])
mse_score = gaussian_func.cost(chosen_params)
print(f'Error for Mu=5, Sigma=1.8: {mse_score}')
```

### Plotting provided data alongside predicted output for selected parameters:

```python
pred_y = gaussian_func.predict(features, chosen_params)

plt.scatter(features, responses, label='feature and true response point')
plt.scatter(features, pred_y, label='feature and predicted response point')
plt.title("Response Values vs. Feature Values", fontsize=15)
plt.xlabel("Feature Values", fontsize=12)
plt.ylabel("Response Values", fontsize=12)
plt.legend()

plt.suptitle(f'Original Data and Predicted Response vars (MSE = {round(mse_score, 5)})', fontsize=18)
plt.legend(loc=2)
plt.ylim(-.02, .45)
plt.tight_layout()
```

### Finding minimum MSE error for refinement pass grid search:

```python
mse_idx = np.argmin(mse)
min_mse = mse[mse_idx]
min_mu = params[mse_idx, 0]
min_sigma = params[mse_idx, 1]
print(f'Minimum MSE score in current parameter space: {min_mse}')
print(f'Coresponding Mu value: {min_mu}')
print(f'Coresponding Sigma value: {min_sigma}')
```

## Experiment 3: “Blind” Grid Search – Multivariate Linear Model
### Loading advertising.csv data:

```python
data_E2 = np.loadtxt(open("advertising.csv", "rb"), delimiter=",", skiprows=1)
data_column_labels = np.loadtxt(open("advertising.csv", "rb"), delimiter=",", dtype='S', max_rows=1)
print("Advertising data shape: " + str(data_E2.shape))
print("Advertising data column labels: " + str(data_column_labels))
```

- #### Identifying response variable and feature variables:
  - After viewing the labels for the Advertising data set, it is clear that the feature variables are the second, third, fourth, and fith columns, and the response variable is the final 'sales' column.

### Instantiating LinearCostFunction using the provided data:

```python
features = data_E2[:, 1:5]
sale_values = data_E2[:, 5]

linear_model = LinearCostFunction(features, sale_values)
```

###  Plotting model predictions versus true response:

```python
params = np.array([0.1, 0.1, 0.1, 0.1])
pred_y = linear_model.predict(features, params)
mse_score = linear_model.cost(params)
plt.scatter(pred_y, sale_values, label='predicted response and true response point')
plt.plot([0, 45], [0, 45], c = 'r', label = 'Line of Identity')
plt.xlabel('Predicted Sale Values')
plt.ylabel('True Sale Values')
plt.title(f'True Sale Values vs Predicted Sale Values (MSE = {round(mse_score, 5)})')
plt.legend()
```

### Performing grid search over 4 dimensional parameter space:

```python
weight = np.linspace(-1, 1, 50)
w0, w1, w2, w3 = np.meshgrid(weight, weight, weight, weight)
params = np.array([w0.flatten(), w1.flatten(), w2.flatten(), w3.flatten()]).T
mse = np.zeros(params[:, 0].shape)

for i in range(mse.size):
    mse[i] = linear_model.cost(params[i])
```

### Finding minimum MSE error for 4 dimensional grid search:

```python
mse_idx = np.argmin(mse)
min_mse = mse[mse_idx]
min_B0 = params[mse_idx, 0]
min_B1 = params[mse_idx, 1]
min_B2 = params[mse_idx, 2]
min_B3 = params[mse_idx, 3]
print(f'Minimum MSE score in current parameter space: {min_mse}')
print(f'Coresponding B0 value: {min_B0}')
print(f'Coresponding B1 value: {min_B1}')
print(f'Coresponding B2 value: {min_B2}')
print(f'Coresponding B3 value: {min_B3}')
```

###  Plotting gird search optimized model predictions versus true response:

```python
pred_y = linear_model.predict(features, params[mse_idx, :])
mse_score = linear_model.cost(params[mse_idx, :])
plt.scatter(pred_y, sale_values, label='predicted response and true response point')
plt.plot([0, 31], [0, 31], c = 'r', label = 'Line of Identity')
plt.xlabel('Predicted Sale Values')
plt.ylabel('True Sale Values')
plt.title(f'True Sale Values vs Predicted Sale Values (MSE = {round(mse_score, 5)})')
plt.legend()
```
